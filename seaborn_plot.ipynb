```python
import seaborn as sns
import pandas as pd
sns.set()

athletes = pd.read_csv('./data/advanced/athletes.csv')
athletes = athletes[athletes['sport'].isin(['wrestling', 'badminton', 'triathlon', 'gymnastics'])]

```

```python
# sns.relplot(data=athletes, x='weight', y='height', col='sport', col_wrap=2,hue='sex', kind='line');
sns.relplot(data=athletes, x='weight', y='height', col='sport', col_wrap=2,hue='sex');
```
